```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

csvpath1 = "Data_Sets/whp_depression_merge.csv"
csvpath2 = "Data_Sets/Country_Coordinates.csv"

whp_dep_df = pd.read_csv(csvpath1)
country_coord_df = pd.read_csv(csvpath2)
```

```python
#remove unnecessary columns

whp_dep_df = pd.DataFrame(whp_dep_df[["Country",
                                      "Happiness Rank 2015",
                                      "Happiness Rank 2016",
                                      "Happiness Rank 2017",
                                      "Happiness Score 2015",
                                      "Happiness Score 2016",
                                      "Happiness Score 2017",
                                      "2015_pop_w_depression",
                                      "2016_pop_w_depression"]])
whp_dep_df.head()
```

```python
whp_dep_df = whp_dep_df.rename(columns ={"2015_pop_w_depression" : "Percent Pop w Depression 2015",
                                         "2016_pop_w_depression" : "Percent Pop w Depression 2016"})
```

```python
country_coord_df = country_coord_df.rename(columns = {"country" : "Country Code",
                                                      "name" : "Country",
                                                      "latitude":"Latitude",
                                                      "longitude":"Longitude"})

country_coord_df = pd.DataFrame(country_coord_df[["Country",
                                                 "Country Code",
                                                 "Latitude",
                                                 "Longitude"]])
country_coord_df.head()
```

```python
merged_w_coord_df = pd.merge(country_coord_df, whp_dep_df, on = "Country")
merged_w_coord_df.head()
```

```python
merged_w_coord_df.count()
```

```python
#calculate average score for 3 years for happiness score, 2 years depression rate, add to df

merged_w_coord_df['Average Happiness Score 2015-2017'] = merged_w_coord_df[['Happiness Score 2015',
                                                                  'Happiness Score 2016',
                                                                  'Happiness Score 2017']].mean(axis=1)
merged_w_coord_df['Average % Pop w Depression 2015-2016'] = merged_w_coord_df[['Percent Pop w Depression 2015',
                                                                  'Percent Pop w Depression 2016']].mean(axis=1)
merged_w_coord_df.head()
```

```python
#export to csv
merged_w_coord_df.to_csv("Data_Sets/merged_whp_dep_coord.csv")
```

```python
#graph latitude in relation to happiness score for average 2015-2017
#check min/max/mean of average happiness score

happy_lat_df = pd.DataFrame(merged_w_coord_df[["Country", "Average Happiness Score 2015-2017", "Latitude"]])

print(happy_lat_df["Average Happiness Score 2015-2017"].max())
print(happy_lat_df["Average Happiness Score 2015-2017"].min())
print(happy_lat_df["Average Happiness Score 2015-2017"].mean())

happy_lat_df.head()
```

```python
#find info on countries scoring 6.5 and over
happiest_df = happy_lat_df.loc[happy_lat_df["Average Happiness Score 2015-2017"] >= 6.5,
                                   ["Latitude", "Country"]]
happiest_df = pd.merge(happiest_df, happy_lat_df, on = "Country")
del happiest_df["Latitude_x"]
happiest_df = happiest_df.rename(columns = {"Latitude_y":"Latitude"})

happiest_df["size"] = pd.DataFrame(happiest_df["Average Happiness Score 2015-2017"]*30)
```

```python
#find info on countries scoring 5.75-6.5
happier_df = happy_lat_df.loc[(happy_lat_df["Average Happiness Score 2015-2017"] >= 5.75) & 
                                      (happy_lat_df["Average Happiness Score 2015-2017"] < 6.5),
                                   ["Latitude", "Country"]]
happier_df = pd.merge(happier_df, happy_lat_df, on = "Country")

del happier_df["Latitude_x"]
happier_df = happier_df.rename(columns = {"Latitude_y":"Latitude"})
len(happier_df)

happier_df["size"] = pd.DataFrame(happier_df["Average Happiness Score 2015-2017"]*30)
```

```python
#find info on countries scoring 5-5.75

meh_df = happy_lat_df.loc[(happy_lat_df["Average Happiness Score 2015-2017"] >= 5) & 
                                      (happy_lat_df["Average Happiness Score 2015-2017"] < 5.75),
                                   ["Latitude", "Country"]]
meh_df = pd.merge(meh_df, happy_lat_df, on = "Country")

del meh_df["Latitude_x"]
meh_df = meh_df.rename(columns = {"Latitude_y":"Latitude"})
len(meh_df)

meh_df["size"] = pd.DataFrame(meh_df["Average Happiness Score 2015-2017"]*30)
```

```python
#find info on countries scoring 4.25-5
sad_df = happy_lat_df.loc[(happy_lat_df["Average Happiness Score 2015-2017"] >= 4.25) & 
                                      (happy_lat_df["Average Happiness Score 2015-2017"] < 5),
                                   ["Latitude", "Country"]]
sad_df = pd.merge(sad_df, happy_lat_df, on = "Country")

del sad_df["Latitude_x"]
sad_df = sad_df.rename(columns = {"Latitude_y":"Latitude"})
len(sad_df)
sad_df["size"] = pd.DataFrame(sad_df["Average Happiness Score 2015-2017"]*30)
```

```python
#find info on countries scoring less than 4.25
saddest_df = happy_lat_df.loc[(happy_lat_df["Average Happiness Score 2015-2017"] < 4.25),
                                   ["Latitude", "Country"]]
saddest_df = pd.merge(saddest_df, happy_lat_df, on = "Country")

del saddest_df["Latitude_x"]
saddest_df = saddest_df.rename(columns = {"Latitude_y":"Latitude"})
len(saddest_df)

saddest_df["size"] = pd.DataFrame(saddest_df["Average Happiness Score 2015-2017"]*30)
```

```python
plt.scatter(happiest_df["Average Happiness Score 2015-2017"], happiest_df["Latitude"],
            facecolors="red", edgecolors="black", s = happiest_df["size"])
plt.scatter(happier_df["Average Happiness Score 2015-2017"], happier_df["Latitude"],
            facecolors="orange", edgecolors="black", s = happier_df["size"])
plt.scatter(meh_df["Average Happiness Score 2015-2017"], meh_df["Latitude"],
            facecolors="yellow", edgecolors="black", s = meh_df["size"])
plt.scatter(sad_df["Average Happiness Score 2015-2017"], sad_df["Latitude"],
            facecolors="lightgreen", edgecolors="black", s = sad_df["size"])
plt.scatter(saddest_df["Average Happiness Score 2015-2017"], saddest_df["Latitude"],
            facecolors="blue", edgecolors="black", s = saddest_df["size"])

plt.grid()

# Set your x and y limits
plt.xlim(1,8)
plt.ylim(-80,80)

# Incorporate the other graph properties
plt.title("Average Happiness of Country by Latitude")
plt.xlabel("World Happiness Report Happiness Score")
plt.ylabel("Latitude")


# Create a legend
legend = plt.legend (('>6.5', '5.75-6.49', '5-5.75', '4.25-4.99', '<4.25'),
                     title = "Happiness Score",
                     scatterpoints = 1,
                     loc = 'lower left')

legend.legendHandles[0]._sizes = [20]
legend.legendHandles[1]._sizes = [20]
legend.legendHandles[2]._sizes = [20]
legend.legendHandles[3]._sizes = [20]
legend.legendHandles[4]._sizes = [20]

plt.grid(True)
plt.axhline(linewidth=4, color='k')


plt.savefig("Images/avg_happiness_by_lat.png")
```

```python
#graph latitude in relation to population with depression for average 2015-2016
#check min/max/mean of average depression

depression_lat_df = pd.DataFrame(merged_w_coord_df[["Country", "Average % Pop w Depression 2015-2016", "Latitude"]])

print(depression_lat_df["Average % Pop w Depression 2015-2016"].max())
print(depression_lat_df["Average % Pop w Depression 2015-2016"].min())
print(depression_lat_df["Average % Pop w Depression 2015-2016"].mean())

depression_lat_df.head()
```

```python
#find info on countries with greater than 4.25% depressed population
most_depressed_df = depression_lat_df.loc[(depression_lat_df["Average % Pop w Depression 2015-2016"] > 4.25),
                                   ["Latitude", "Country"]]
most_depressed_df = pd.merge(most_depressed_df, depression_lat_df, on = "Country")

del most_depressed_df["Latitude_x"]
most_depressed_df = most_depressed_df.rename(columns = {"Latitude_y":"Latitude"})
print(len(most_depressed_df))

most_depressed_df.head()

most_depressed_df["size"] = pd.DataFrame(most_depressed_df["Average % Pop w Depression 2015-2016"]*30)
```

```python
#find info on countries with 4-4.25% depressed population
kinda_depressed_df = depression_lat_df.loc[(depression_lat_df["Average % Pop w Depression 2015-2016"] > 4) &
                                           (depression_lat_df["Average % Pop w Depression 2015-2016"] <=4.25),
                                   ["Latitude", "Country"]]
kinda_depressed_df = pd.merge(kinda_depressed_df, depression_lat_df, on = "Country")

del kinda_depressed_df["Latitude_x"]
kinda_depressed_df = kinda_depressed_df.rename(columns = {"Latitude_y":"Latitude"})
print(len(kinda_depressed_df))

kinda_depressed_df.head()

kinda_depressed_df["size"] = pd.DataFrame(kinda_depressed_df["Average % Pop w Depression 2015-2016"]*30)
```

```python
#find info on countries with 3.75-4% depressed population
depressed_df = depression_lat_df.loc[(depression_lat_df["Average % Pop w Depression 2015-2016"] > 3.75) &
                                           (depression_lat_df["Average % Pop w Depression 2015-2016"] <=4),
                                   ["Latitude", "Country"]]
depressed_df = pd.merge(depressed_df, depression_lat_df, on = "Country")

del depressed_df["Latitude_x"]
depressed_df = depressed_df.rename(columns = {"Latitude_y":"Latitude"})
print(len(depressed_df))

depressed_df.head()

depressed_df["size"] = pd.DataFrame(depressed_df["Average % Pop w Depression 2015-2016"]*30)
```

```python
#find info on countries with 3.75-3.75% depressed population
lite_depressed_df = depression_lat_df.loc[(depression_lat_df["Average % Pop w Depression 2015-2016"] > 3.25) &
                                           (depression_lat_df["Average % Pop w Depression 2015-2016"] <=3.75),
                                   ["Latitude", "Country"]]
lite_depressed_df = pd.merge(lite_depressed_df, depression_lat_df, on = "Country")

del lite_depressed_df["Latitude_x"]
lite_depressed_df = lite_depressed_df.rename(columns = {"Latitude_y":"Latitude"})
print(len(lite_depressed_df))

lite_depressed_df.head()

lite_depressed_df["size"] = pd.DataFrame(lite_depressed_df["Average % Pop w Depression 2015-2016"]*30)
```

```python
#find info on countries with less than 3.25% depressed population
notso_depressed_df = depression_lat_df.loc[(depression_lat_df["Average % Pop w Depression 2015-2016"] <= 3.25),
                                   ["Latitude", "Country"]]
notso_depressed_df = pd.merge(notso_depressed_df, depression_lat_df, on = "Country")

del notso_depressed_df["Latitude_x"]
notso_depressed_df = notso_depressed_df.rename(columns = {"Latitude_y":"Latitude"})
print(len(notso_depressed_df))

notso_depressed_df.head()

notso_depressed_df["size"] = pd.DataFrame(notso_depressed_df["Average % Pop w Depression 2015-2016"]*30)
```

```python
plt.scatter(most_depressed_df["Average % Pop w Depression 2015-2016"], most_depressed_df["Latitude"],
            facecolors="black", edgecolors="black", s = most_depressed_df["size"])
plt.scatter(kinda_depressed_df["Average % Pop w Depression 2015-2016"], kinda_depressed_df["Latitude"],
            facecolors="darkblue", edgecolors="black", s = kinda_depressed_df["size"])
plt.scatter(depressed_df["Average % Pop w Depression 2015-2016"], depressed_df["Latitude"],
            facecolors="blue", edgecolors="black", s = depressed_df["size"])
plt.scatter(lite_depressed_df["Average % Pop w Depression 2015-2016"], lite_depressed_df["Latitude"],
            facecolors="lightblue", edgecolors="black", s = lite_depressed_df["size"])
plt.scatter(notso_depressed_df["Average % Pop w Depression 2015-2016"], notso_depressed_df["Latitude"],
            facecolors="cyan", edgecolors="black", s = notso_depressed_df["size"])

# Set your x and y limits
plt.xlim(0,6)
plt.ylim(-80,80)

# Incorporate the other graph properties
plt.title("Average % Depressed Population of Country by Latitude")
plt.xlabel("% Depressed Population")
plt.ylabel("Latitude")


# Create a legend
legend = plt.legend (('>4.25%', '4-4.25%', '3.75-4%', '3.25-3.75%', '<3.25'),
                     title = "% Depressed Pop.",
                     scatterpoints = 1,
                     loc = 'best')

legend.legendHandles[0]._sizes = [30]
legend.legendHandles[1]._sizes = [30]
legend.legendHandles[2]._sizes = [30]
legend.legendHandles[3]._sizes = [30]
legend.legendHandles[4]._sizes = [30]

plt.grid(True)
plt.axhline(linewidth=4, color='r')

plt.savefig("Images/avg_depression_by_lat.png")
```
