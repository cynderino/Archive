```python
# Dependencies
import pandas as pd
import numpy as np
```

```python
# Name of the CSV file
Happiness15 = "2015_World_Happiness_Report.csv"
Happiness16 = "2016_World_Happiness_Report.csv"
Happiness17 = "2017_World_Happiness_Report.csv"
```

```python
# The correct encoding must be used to read the CSV in pandas
H15_df = pd.read_csv(Happiness15)
H16_df = pd.read_csv(Happiness16)
H17_df = pd.read_csv(Happiness17)
```

```python
# Preview of the DataFrame
HRS15_df = pd.DataFrame(H15_df[["Country","Happiness Rank","Happiness Score"]])
HRS15_df = HRS15_df.rename(columns={"Happiness Rank": "Happiness Rank 2015","Happiness Score":"Happiness Score 2015"})
HRS15_df.head()
# HRS15_df.count()
# HRS15_df.dtypes
```

```python
# Preview of the DataFrame
HRS16_df = pd.DataFrame(H16_df[["Country","Happiness Rank","Happiness Score"]])
HRS16_df = HRS16_df.rename(columns={"Happiness Rank": "Happiness Rank 2016","Happiness Score":"Happiness Score 2016"})
HRS16_df.head()
# HRS16_df.count()
# HRS16_df.dtypes
```

```python
# Preview of the DataFrame
HRS17_df = pd.DataFrame(H17_df[["Country","Happiness.Rank","Happiness.Score"]])
HRS17_df = HRS17_df.rename(columns={"Happiness.Rank": "Happiness Rank 2017","Happiness.Score":"Happiness Score 2017"})
HRS17_df.head()
# HRS17_df.count()
# HRS17_df.dtypes
```

```python
# Merge the DataFrame
happy1_df = pd.merge(HRS15_df, HRS16_df, on="Country", how="inner")
happy1_df.head()
```

```python
# Merge the DataFrame
world_happy_df = pd.merge(happy1_df, HRS17_df, on="Country", how="inner")
world_happy_df
```

```python
#Save File
world_happy_df.to_csv("data_sets/world_happy.csv")
```
